```python
import json
path="usa.txt"
records=[]
for line in open(path) :
        records.append(json.loads(line))
print(records[0])
print()
print(records[0]['a'])
print()
print(records[0]['tz'])
time_zones=[]
for rec in records :
    if 'tz' in rec :
        time_zones.append(rec['tz'] )
print()
print(time_zones)
print(time_zones[:10])
```

```python
def get_counts(time_zones):
    counts={}
    for i in time_zones :
        if i in counts :
            counts[i] += 1
        else :
            counts[i]=1
    return counts 
#def get_counts(time_zones):
   # counts={}
   # for i in time_zones :
    #    counts[i]=time_zones.count(i)
   # return counts
dic_counts=get_counts(time_zones)
print(dic_counts)
print()
print(dic_counts['America/New_York'])
print(len(dic_counts))
#top 10
def top_counts(dic_counts):
    top_10 = [(count, tz) for tz, count in dic_counts.items()]
    #top_10=[]
    #for tz, count in dic_counts.items() :
    #top_10.append(count,tz)
    top_10.sort()
    top_10_n=[]
    for i in range(len(top_10)-1,len(top_10)-11,-1):
        top_10_n.append(top_10[i])
    return top_10_n    
    
dic_top_10=top_counts(dic_counts)
print(dic_top_10)
```

```python
from collections import defaultdict
from collections import Counter
def get_counts(time_zones):
    counts = defaultdict(int) # values will initialize to 0
    for x in time_zones:
        counts[x] += 1
    return counts
print(get_counts(time_zones))
print()
print(get_counts(time_zones)['America/New_York'])
count=Counter(get_counts(time_zones))
#print(count)
print(count.most_common(10))
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
cframe=frame[frame.notnull()]
print(cframe)
print()
windows=np.where(cframe['a'].str.contains('Windows'),'Windows', 'Not Windows')
print(windows)
print()
print(windows[:5])
print()
by_tz_os = cframe.groupby(['tz', windows])
agg_counts = by_tz_os.size().unstack().fillna(0)
print(agg_counts)
print()
print(agg_counts[:10])
```

```python
indexer=agg_counts.sum(1).argsort()
print(indexer)
print()
print(indexer[:10])
```

```python
count_subset = agg_counts.take(indexer)
print(count_subset)
count_subset = agg_counts.take(indexer)[-10:]
print(count_subset)
```

```python
count_subset.plot(kind='barh',stacked=True)
```

```python
normed_subset=count_subset.div(count_subset.sum(1),axis=0)
normed_subset.plot(kind='barh',stacked=True)
```
