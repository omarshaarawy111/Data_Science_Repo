```python
def greet(name):
    print('Hello {}'.format(name))
    #print(f'Hello {name}')
    #print(f"Hello {name}")
def even_odd(number):
    if number % 2 == 0 :
        return 'even'
    else :
        return 'odd'
def is_even(number):
    if number % 2 == 0 :
        return True
    else :
        return False 
    
def is_odd(number):
    if number % 2 != 0 :
        return True
    else :
        return False   
    
def area(number):
    return 3.14*pow(number,2)
def add(number=0,number_1=1):
    return number + number_1
def is_prime(number):
    counter=0
    for i in range(2,number):
        if number % i  !=0 :
            counter +=1
            
    if counter == number-2 :
        return True
    if counter <= number :
        return False

def is_prime_1(number):
    counter=0
    for i in range(2,number):
        if number % i  == 0 :
            counter +=1
            break;
            
    if counter > 0 :
        return False
    else  :
        return True  
    
def is_prime_2(number):
    for i in range(2,number):
        if number % i  == 0 :
            return False
            break;
            
    else :
        return True
    
def is_prime_3(number):
    for i in range(2,number):
        if number % i  == 0 :
            print('False')
            break;
            
    else :
        print('True') 
        
def is_prime_4(number):
    for i in range(2,number):
        if number % i  == 0 :
            return 'False'
            break;
            
    else :
        return 'True'        
    
    
def take():
    name=input("enter name : ")
    number=int(input("enter number  : "))
    number_1=int(input("enter number : "))
    print(even_odd(number))
    print(is_even(number))
    print(is_odd(number))
    print(area(number))
    print(add(number,number_1))
    print(add('one','two'))
    print(add())
    print(add(1))
    greet(name)
    print(type(add))
    print(is_prime(number))
    print(is_prime_1(number))
    print(is_prime_2(number))
    is_prime_3(number)
    print(is_prime_4(number))
    
take()    
```

```python
from functools import reduce

def odd(n):
    return n % 2 !=0

def cube(n):
    return n ** 3

def add(a,b):
    return a+b

def take():
    lst=list(input("enter list : ").split())
    for x in range(len(lst)) :
        lst[x]=int(lst[x])
    odd_list=list(filter(odd,lst))  
    sum_list=list(map(cube,odd_list))
    print("the sum is = ",reduce(add,sum_list))   
    
take()    
```

```python
from functools import reduce

def odd(n):
    return n % 2 !=0

def cube(n):
    return n ** 3

def greater(a,b):
    if a > b :
        return a
    else :
        return b

def take():
    lst=list(range(1,20))
    for x in range(len(lst)) :
        lst[x]=int(lst[x])
    odd_list=list(filter(odd,lst)) 
    sum_list=list(map(cube,odd_list))
    print("the max is = ",reduce(greater,sum_list))   
    
take()
```
