```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
url = "https://www.scrapethissite.com/pages/forms/?page_num=1"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
team_package=soup.find('table',attrs={'class':'table'}).find_all('tr')
team_package[1:]
```

```python
teams=[]
for r in team_package[1:] :
    team_name=r.find('td',attrs={'class':'name'}).get_text().strip().split(" ")[-1]
    year=int(r.find('td',attrs={'class':'year'}).get_text().strip().split(" ")[-1])
    wins=int(r.find('td',attrs={'class':'wins'}).get_text())
    losses=int(r.find('td',attrs={'class':'losses'}).get_text())
    win_per=float(r.find('td',attrs={'class':'pct'}).get_text())
    gf=int(r.find('td',attrs={'class':'gf'}).get_text())
    ga=int(r.find('td',attrs={'class':'ga'}).get_text())
    pm=int(r.find('td',attrs={'class':'diff'}).get_text())
    team_data={'team Name':team_name,'year':year,'wins':wins,'losses':losses,'win %':win_per,'goals for':gf,'goals agianst':ga,'plus minis':pm}
    print(team_data)
    teams.append(team_data)
```

```python
teams
```

```python
with open('team.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['team Name', 'year', 'wins', 'losses','win %','goals for','goals agianst','plus minis'])
    writer.writeheader()
    for t in teams:
        writer.writerow(t)  
team_table=pd.read_csv('team.csv')
team_table
```
