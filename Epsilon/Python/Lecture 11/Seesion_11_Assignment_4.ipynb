```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
country_package=soup.find_all('div',attrs={'class':'col-md-4 country'})
country_package
```

```python
countries=[]
for c in country_package :
    country_name=c.find('h3',attrs={'class':'country-name'}).get_text().strip().split(" ")[-1]
    country_capital=c.find('span',attrs={'class':'country-capital'}).get_text()
    country_population=int(c.find('span',attrs={'class':'country-population'}).get_text())
    country_area=float(c.find('span',attrs={'class':'country-area'}).get_text())
    country_data={'country name':country_name,'country capital':country_capital,'country population':country_population,'country area':country_area}
    print(country_data)
    countries.append(country_data)
```

```python
countries
```

```python
with open('country.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['country name', 'country capital', 'country population', 'country area'])
    writer.writeheader()
    for co in countries:
        writer.writerow(co)  
country_table=pd.read_csv('country.csv')
country_table
```
