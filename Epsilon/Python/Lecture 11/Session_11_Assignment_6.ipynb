```python
import requests
import csv
import pandas as pd
url='https://jsonplaceholder.typicode.com/users'
response=requests.get(url)
response
```

```python
users=response.json()
users
```

```python
with open('users.csv','w',newline='',encoding='utf-8') as f :
    writer=csv.writer(f)
    headers=['name','username','email','phone', 'website', 'company name','street','suite','city','zipcode','geo_lat','geo_lng']
    writer.writerow(headers)
    for user in users:
        name=user['name']
        user_name=user['username']
        email=user['email']
        phone=user['phone']
        website=user['website']
        company_name=user['company']['name']
        street=user['address']['street']
        suite=user['address']['suite']
        city=user['address']['city']
        zip_code=user['address']['zipcode']
        geo_lat=user['address']['geo']['lat']
        geo_lng=user['address']['geo']['lng']
        data=[name,user_name,email,phone,website,company_name,street,suite,city,zip_code,geo_lat,geo_lng]
        writer.writerow(data)
        
        

        
```
