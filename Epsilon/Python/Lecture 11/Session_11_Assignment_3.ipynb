```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
rates={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
response = requests.get('http://books.toscrape.com')
soup = BeautifulSoup(response.content, 'html.parser')
catogories_link=soup.find('div',attrs={'class':'side_categories'}).find('ul').find('ul').find_all('a')
catogories_link
```

```python
books=[]
for c in catogories_link:
    category_name=c.get_text().strip()
    category_url = f"http://books.toscrape.com/{c.get('href')}"
    print(f"for category {category_name} \n: {category_url}")
    response = requests.get(category_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    article_all=soup.find_all('article',attrs={'class':'product_pod'})
    for a in article_all :
        price=float(a.find('p',attrs={'class':'price_color'}).get_text().split('£')[1])
        title=a.find('h3').find('a').get_text()
        rate=rates[a.find('p').get('class')[1]]
        book_data={'title':title,'price':price,'rate':rate,'category name':category_name}
        print(book_data)
        books.append(book_data)
        
    
```

```python
books
```

```python
with open('books.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['title', 'price', 'rate', 'category name'])
    writer.writeheader()
    for book_data in books:
        writer.writerow(book_data)                                  
                                         
    
```
