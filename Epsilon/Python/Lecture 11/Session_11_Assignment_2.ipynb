```python
#for today
import requests
from bs4 import BeautifulSoup
import pandas as pd
url='https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
periods_get=soup.find_all('p',attrs={'class':'period-name'})
periods=[]
periods_get  
for p in periods_get:
    periods.append(p.get_text())
periods  
title_get=soup.find_all('img',attrs={'class':'forecast-icon'})
title=[]
for t in title_get:
    title.append(t['title'])
title  
short_desc_get=soup.find_all('p',attrs={'class':'short-desc'})
short_desc=[]
for s in short_desc_get :
    short_desc.append(s.get_text())
short_desc 
temp_get=soup.select('.forecast-tombstone .temp ')
temp=[]
for t in temp_get:
    temp.append(t.get_text())
temp   
weather = pd.DataFrame({'period':periods,'title':title,'short desc':short_desc,'temp':temp})
f=weather['temp'].str.extract('(\d+)')
weather['temp_in_f']=f.astype('int')
weather['temp_in_c'] = (weather['temp_in_f']-32)*(5/9)
weather[0:1]
```
